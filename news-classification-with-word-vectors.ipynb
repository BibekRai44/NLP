```python
import pandas as pd
```

```python
df=pd.read_csv("/kaggle/input/fake-real-news-dataset/Fake_Real_Data.csv")
df.head()
```

```python
df['label_num']=df['label'].map({'Fake':0,'Real':1})
df.head()
```

```python
import spacy
nlp=spacy.load("en_core_web_lg")
```

```python
df['vector_relation']=df['Text'].apply(lambda text:nlp(text).vector)
```

```python
df.head()
```

```python
x=df.vector_relation.values
y=df.label_num
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,stratify=df['label_num'])
```

```python
import numpy as np
X_train_stacked=np.stack(X_train)
X_test_stacked=np.stack(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
from sklearn.preprocessing import MinMaxScaler
model=MinMaxScaler()
scaled_train_embed = model.fit_transform(X_train_stacked)
scaled_test_embed = model.transform(X_test_stacked)
clf=MultinomialNB()
clf.fit(scaled_train_embed,y_train)
y_pred=clf.predict(scaled_test_embed)
print(classification_report(y_test,y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=5,metric='euclidean')
model.fit(scaled_train_embed,y_train)
y_pred=model.predict(scaled_test_embed)
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
from matplotlib import pyplot as plt
%matplotlib inline
plt.figure(figsize=(6,6))
import seaborn as sns
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Prediction')
plt.ylabel('Truth')
```
