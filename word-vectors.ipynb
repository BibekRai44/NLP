```python
import spacy
```

```python
nlp=spacy.load("en_core_web_lg")
doc=nlp("America is a shitty country and is approved by bibek")
for token in doc:
    print(token,token.has_vector)
```

```python
doc=nlp("profit loss gain bussiness")
for token in doc:
    print(token.text,"|",token.similarity(doc))
```
